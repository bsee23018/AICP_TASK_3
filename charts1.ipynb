# Question 2

```python
x_axxis= [10,20,20,40,30,10]
y_axxis= [10,40,20,10,30,30]
base_line= [2,4,6,8,10,12]
plt.title('Display Meter')
plt.plot(base_line , x_axxis,label = 'blue' , marker = 'o' , markerfacecolor = 'blue')
plt.plot(base_line , y_axxis,label = 'yellow', marker = 'o', markerfacecolor = 'yellow')
plt.xlabel('x_axxis')
plt.ylabel('y_axxis')
plt.legend()
plt.show()
```

# Question 5

```python
plt.title('Comaparison of colors')
p1 =[4,2,4,2,2]
p2 =[8,3,7,6,4]
p3 =[5,4,4,4,3]
p4 =[7,2,7,8,3]
p5 =[6,6,8,6,2]
x_side=[2,4,6,8,10]
e = 0.1
p1_bar = np.arange(len(x_side))
p2_bar = [i + e for i in p1_bar]
p3_bar = [i + e for i in p2_bar]
p4_bar = [i + e for i in p3_bar]
p5_bar = [i + e for i in p4_bar]
plt.bar(p1_bar,p1,width=e , label = 'a',color= 'darkblue')
plt.bar(p2_bar,p2,width=e , label = 'b',color= 'darkgreen')
plt.bar(p3_bar,p3,width=e , label = 'c',color= 'red')
plt.bar(p4_bar,p4,width=e , label = 'd',color= 'skyblue')
plt.bar(p5_bar,p5,width=e , label = 'e',color= 'purple')
plt.xticks(p1_bar +e+e/2,x_side)



plt.legend()
plt.show()
```

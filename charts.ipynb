# Question 2

```python
x_axxis= [10,20,20,40,30,10]
y_axxis= [10,40,20,10,30,30]
base_line= [2,4,6,8,10,12]
plt.title('Display Meter')
plt.plot(base_line , x_axxis,label = 'blue' , marker = 'o' , markerfacecolor = 'blue')
plt.plot(base_line , y_axxis,label = 'yellow', marker = 'o', markerfacecolor = 'yellow')
plt.xlabel('x_axxis')
plt.ylabel('y_axxis')
plt.legend()
plt.show()
```
